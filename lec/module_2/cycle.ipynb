# Цикл for

Цикл for - это последовательные очередный цикл перебора значений. В этом цикле, мы работаем только с итерирумыми структурами данных (например, списки, кортежи, строки). Цикл for используется для того, чтобы выполнить блок кода несколько раз.


````python
lst = [2,3,4,5,6,7,8,9]

for val in lst: 
    print(f'Значение: {val=}')

print(f'{val=}')
````

### val - в данном случае это временная переменная, которая хранит промежуточный результат цикла

Если мы используюем в итерируемом объекте, связную структуру [(1,2), (2,3)] с точным количеством элементов внутри вложенной структуры то
мы можем использовать несколько временных переменных.

Такой формат записи, когда мы "распоковываем" внутренние элементы под определенные временные переменные, называется распаковкой


````python
coords = [(2,3), (3,4)]
#ВАЖНО: Это работает только в том случае если длина вложенной структуры равна одинаковому значению на протяжении всей основной структуры(coords)

# coords = [(2,3), (3,4), (4,5,6)] # Эта структура работать не будет, так как длина последнего вложенного элемента, равна 3 

for x,y in coords: 
    print(f'Значение: {x=} {y=}')
````

# Пример 1

Проход по числам списка с последующим добавлением чётных чисел в список lst2

````python
lst = [2,3,4,5,6,7,8,9]
lst2 = []

for val in lst: 
    print(f'Значение: {val=}')

    if val % 2 == 0:
        print(f'Число {val} - чётное')
        lst2.append(val)

print(f"{lst2=}")
````

Работа с циклом for по значениям индекса

Функция `range(y)` - это функция которая позволяет составить диапазон чисел от 0 до y-1 при стандартном обращении

В расширенной форме:

`range(x, y, step)`- где x - это начало(по умолчанию 0); y - это конец диапазона; step - шаг между элементами диапазона, по умолчанию 1

Функция `len(*объект*)` - это стандартная функция для получения количества элементов в объекте(длины объекта) [от слова len\[ght\](длина)]



````python
lst = [2,3,4,5,6,7,8,9,10]

print(f"{range(10)=}")
print(f"{range(5, 20)=}") 
# range() это объект "ленивого вычисления" поэтому он не отображается в чистов виде если его не обработать
print(f"{list(range(5, 20))=}")
print(f"{list(range(2, 20, 2))=}")

print(f'{lst=}| len = {len(lst)}')

for i in range(len(lst)):
    print(f"lst[{i}]={lst[i]}")
````

`enumerate(*объект*)`*- стандартная функция Python, которая позволяет перебрать элементы объекта и их индекс 


````python
# Предположим что вам нужно пройтись по списку из чисел от 1 до 100 и выводить только те значения и их индекс которые кратны 4 и 6

import random


lst = list(range(1,101))
random.shuffle(lst)# перетасуем значение в списке чтобы они были неупорядоченные

# в при использовании стандартного подхода ваш код выглядит вот так
for i in range(len(lst)):
    if lst[i] % 4 == 0 and lst[i] % 6 == 0:
        print(f"lst[{i}]={lst[i]}")

# Если бы использовали enumerate
print('enumerate')
for i, val in enumerate(lst):
    if val % 4 == 0 and val % 6 == 0:
        print(f'lst[{i}]={val}')
````

Посмотрим как выглядит enumerate в его чиистом представлении:


````python
print(enumerate(lst)) # если его в чистом виде то мы получим лишь лениво обработанный объект
# Поэтому вы делаем вызов с методом list() для получения списка значений enumerate
print(list(enumerate(lst)))
````

Таким образом мы можем заключить что enumerate - составляет пары (индекс, значение) для конкретного итерируемого объекта

# Как работать в циклах со словарями ?

````python
dct = {
    'one':1, 
    'two':[2,3], 
    'three':(3,4,5), 
    'four': {
        'five':5, 
        'six':6
        } 
    }
# Попробуем поработать со словарем как мы ранее работали со списком или множеством
for val in dct:
    print(val)
print('-'*20)

# После запуска мы видим что вывелись только ключи
# Соответственно правильным подходом для вывода значений будет
for key in dct:
    print(f"{key} = {dct[key]}")

print('-'*20)
# Если есть такая потребность в использовании конкретики, мы можем воспользоваться базовыми методами словаря
for val in dct.values():
    print(f"{val=}")

print('-'*20)
for key, value in dct.items():
    print(f"{key} = {value}")

````

# Цикл while

Цикл - это конструкция для повторения кода какое то количество раз

Конечный цикл - цикл с явным конечным условием, т.е. такое условие горантированно закончится 

Общая конструкция на человеческом языке звучит как
```
ПОКА *условие* => ИСТИНА:
    *Делай(код)*```

Поэтому в случае если условие никак не продвигается цикл может стать бесконечно работающим

````python
cnt = 1

while cnt <= 10:
    if cnt % 2 == 0:
        print(f'Текущий элемент счёта: {cnt=} - чётное')
    else:
        print(f'Текущий элемент счёта: {cnt=} - нечётное')
    cnt += 1 # Обязательно изменяем значение которое влияет на условие в процессе работы
else:
    print('Цикл завершен') # Всегда выполняется после завершения цикла как while так и for




````

Хоть цикл while и не может похвастаться таким разнообразием использования специальных методов и функций как цикл for. Но зато с помощью этого цикла, мы можем создавать абсолютно уникальное поведение цикла-повтора

## Бесконечные циклы - это циклы (обычно while) условие не может быть выполнено до конца

А как нам завершать бесконечные циклы ? 

-- Все достаточно просто, для этого у нас есть специальные конструкции break и continue.

- break - это оператор завершения или "обрыва"(прервывания) цикла
- continue - это оператор пропуска текущего шага цикла и переход к следующему шагу

Пользователь вводит произвольное количество чисел, до числа 0, как только пользователь вводит число 0 - цикл заканчивается. Цель программы найти чётные числа кратные 4 и 6 из чисел которые ввёл пользователь.

````python
result_lst = []


while True:
    x = int(input('Введите число: '))
    if x == 0:
        print('Цикл завершен')
        break
    
    if x % 2 == 0 and (x % 4 == 0 and x % 6 == 0):
        print(f'Чётное число кратное 4 и 6: {x}')
        result_lst.append(x)

    

print(f'Результат: {result_lst=}')


````

Пользователь вводит произвольное количество чисел, до числа 0, как только пользователь вводит число 0 - цикл заканчивается. Цель программы найти чётные числа кратные 4 и 6 из чисел которые ввёл пользователь. И любые числа кратные 10 пользователь пропускает

````python
result_lst = []


while True:
    x = int(input('Введите число: '))
    
    if x == 0:
        print('Цикл завершен')
        break

    if x % 10 == 0:
        print(f'Число {x} пропущено.')
        continue
    
    if x % 2 == 0 and (x % 4 == 0 and x % 6 == 0):
        print(f'Чётное число кратное 4 и 6: {x}')
        result_lst.append(x)

    

print(f'Результат: {result_lst=}')


````
