#  

`[dict].update([dict2])` -    `[dict2]`  `[dict]`.  -    `[dict]`,    \  -    `[dict2]`

````python
print(f'{dct=} | {dct2=}')

dct2['key_2523425'] = '  dct2'
print(f'{dct=} | {dct2=}')

print(dct.update(dct2)) #   ,   
print(f'{dct=}')
````

`[dict].pop(<key>)` -     

````python
print(dct)
# print(dct.pop('key1'))
val = dct.pop('key1')
print(f'{dct=} | {val=}')
````

`[dict].popitem()` -  -  ,     

````python
print(dct)
# print(dct.popitem())
val = dct.popitem()
print(f'{dct=} | {val=}')
````

##         `[ (1, 1), ..., (N, N)  ]`

`[dict].get(<key>, *<answer>  . None))` -       `<key>`,    ,   `None`    

````python
print(f'{dct=}')
print(f'{dct.get("key_2523425")=}')
print(f'{dct.get("key_2")=}')
print(f'{dct.get("key_2", "  ")=}') #        
print(f'{dct.get("key_2", 1)=}') #        
````

`[dict].setdefault(<>, *<>  . None)` -      ,       `[dict]` 

````python
print(f'{dct=}')
print(f"{dct.setdefault('key_2')}")
print(f'{dct=}')
print(f"{dct.setdefault('key_3', ['', '',''])}")
print(f'{dct=}')
print(f"{dct.setdefault('key_2523425', '  ?')}")
print(f'{dct=}')
````

`[dict].clear()` -   

````python
print(f'{dct2=}')
dct2.clear()
print(f'{dct2=}')
````

`[dict].values()` -    ,     (dict_values)

````python
print(f'{dct=}')
print(f'{dct.values()=}')
# dct.values()[0] # ,        ,       
# ,        ,         ,      
print(f'{list(dct.values())[0]=}')

lst_of_values = list(dct.values())
print(f'{lst_of_values[0]=}')
````

`[dict].keys()` -    ,     (dict_keys)

````python
print(f'{dct=}')
print(f'{dct.keys()=}')
#     dict_keys       dict_values
print(f'{list(dct.keys())[0]=}')

lst_of_keys = list(dct.keys())
print(f'{lst_of_keys[0]=}')
````
