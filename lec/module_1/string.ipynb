# Строки - `[str]` - String 
# и индексация

Строки -- это набор(массив) символов заключенный в ковычки


Свойства: 
- Неизменямость
- Итерируемость
- Индексировуемость(нумерованность)


````python
print('Hello" world')
print("Hello' world")
# print('Hello' world')
print('Hello\' world') # \ - (backslash) - обратный слэш
````

Индексация - процесс обращения к конкретному объету внутри итерируемого объекта. Любое обращение к индексам работает через `*объект*[*индекс*]`

Итерируемый объект - объект который хранит в себе последовательность объетов (т.е. <u>строка</u> - последовательность букв заключенных в ковычки)

````python
print("Hello world")
print("Hello world"[4]) # обычный индекс, начинается от 0
print("Hello world"[-1]) # обратный индекс, начинается с -1
# print("Hello world"[11]) # Ошибка, выход за пределы индексов строки
````

Индексовый срез - когда мы задаем начальный индекс и конечный индекс через `:`. И чтение идет от начального индекса <u>включительно</u> до конечного <u>исключительно</u>

````python
print("Hello world"[0:4])
print("Hello world"[0:5])
print("Hello world"[:5])
print("Hello world"[6:])
print("Hello world"[6:])
print("Hello world"[6:100])
print("Hello world"[50:100])
````

Обратное индексирование - обратный процесс индексации, когда мы идем не с лева-направо, справа-налево. Соответвенно, индекс начинает тоже обратный отсчёт от 0, т.е. -1, -2, ... , -N *длина объекта*

````python
print("STRING"[-6])
print("STRING"[-6:-3])
print("STRING"[-3:-6]) # Ничего не получим тк шаг по умолчанию идет всегда вправо
print("STRING"[-3:-6:-1]) # Но получим результат, если зададим шаг -1
print("STRING"[-3:-6:-2])
print("STRING"[-3::-1])
print("STRING"[::]) # Если мы ничего не ставим в значениях стара, конца или шага, то они будут взяты по умолчанию
print("STRING"[::-1])
````

# формат индексации `[*старт = 0*:*конец = len*:*шаг* = 1]`

# Форматирование
- это процесс подстановки значение в строку

1 метод форматирования - .format()

````python
name = 'Bob'
age = 20

greetin_text = "Привет {}. Ваш возраст: {}"

print(greetin_text)
print(greetin_text.format(name, age))
# print(greetin_text.format(name))
# print(greetin_text.format(name, age, 5))
# print(greetin_text.format(name)) # ошибка, потому не соответствует требуемому количеству переменных для строки
greetin_text = "Привет {1}. Ваш возраст: {0}"
print(greetin_text.format(name, age))
greetin_text = "Привет {}. Ваш возраст: {:.2f}"
print(greetin_text.format(name, age))
````

2 способ форматирования: форматирование через %

````python
name = 'Bob'
age = 20

greetin_text = "Привет %s. Ваш возраст: %s"

print(greetin_text)

# print(greetin_text % name) # 1 значение передано вместо 2
# print( greetin_text % name, age ) #  greetin_text % name | age

print( greetin_text % (name, age) )
print( greetin_text % (name, age) )
````

%s - s(tring) - ячейка для любых типов данных, поддерживающее перевод в строковый тип данных

%d - d(igit) - ячейка для целочислвенных значений

%f - f(loat) - ячейка для вещественных значений


````python
name = 'Bob'
age = 20
height = 1.955

greetin_text = "Привет %s. Ваш возраст: %s лет. Ваш рост: %s м."
print( greetin_text % (name, age, height) )

greetin_text = "Привет %s. Ваш возраст: %d лет. Ваш рост: %f м."
print( greetin_text % (name, age, height) )

greetin_text = "Привет %s. Ваш возраст: %d лет. Ваш рост: %.2f м."
print( greetin_text % (name, age, height) )

greetin_text = "Привет %s. Ваш возраст: %d лет. Ваш рост: %.2f м."
print( greetin_text % (name, height, age ) )

greetin_text = "Привет %s. Ваш возраст: %03d лет. Ваш рост: %.2f м."
print( greetin_text % (name, age, height) )

greetin_text = "Привет (%10s). Ваш возраст: (%10d) лет. Ваш рост: (%10.2f) м."
print( greetin_text % (name, age, height) )

greetin_text = "Привет (%-10s). Ваш возраст: (%-10d) лет. Ваш рост: (%-10.2f) м."
print( greetin_text % (name, age, height) )
````

3 способ форматирования - f-string - f-строка

````python
name = 'Bob'
age = 20.9
height = 1.955

greetin_text = f"Привет {name}. Ваш возраст: {age} лет. Ваш рост: {height} м. | ({4+4})"
print(greetin_text)

age = 20# для :d в таком способе важно чтобы тип данных переменной был int
greetin_text = f"Привет {name}. Ваш возраст: {age:d} лет. Ваш рост: {height:.2f} м."
print(greetin_text)

greetin_text = f"Привет ({name:10s}). Ваш возраст: ({age:10d}) лет. Ваш рост: ({height:10.2f}) м."
print(greetin_text) # баг, со строкой, перенос происходит слева-направо при положительном числе 10 перед s (должно быть наоборот)

# -10 для строки вообще не работает
# greetin_text = f"Привет ({name:-10s}). Ваш возраст: ({age:10d}) лет. Ваш рост: ({height:10.2f}) м."
# print(greetin_text) # баг, со строкой, перенос происходит слева-направо при положительном числе 10 перед s (должно быть наоборот)


# debug вывод переменоой
print(f'Данные: {name=} ')
print(f'Данные: {name.upper()=} ')
````

# Методы строк

Методы работы с регистром строк: 
- upper() - перевести все буква в ВЕРХНИЙ регистр.
- lower() - перевести все буква в нижний регистр.
- capitalize() - Первая буква заглавная, все последующие буквы маленькие.

````python
print(f'{name=}')
print(f'{name.lower()=}') # вот так результат не сохраняется
print(f'{name.upper()=}') # вот так результат не сохраняется
print(f'{name=}') 

# Можно сохранить его вручную 
s = name.lower()
print(f'{s=}')

s = name.upper()
print(f'{s=}')

# или переприсвоить значение
print(f'{name=}') 
name = name.lower()
print(f'{name=}') 
print(f'{name.capitalize()=}') 
name = name.capitalize()
print(f'{name=}')
````

`[str].replace(<old>, <new>)` - Заменяет все совпавшие значения в строке `[str]` из простроки `<old>` на строку `<new>`

````python
text = "Какой то там текст то"

print(f'{text=}')

print(f"{text.replace('то', '')=}") # Если выставить, 3-тий аргумент(количество замен) в replace, то замена происходит только определенное количество раз
print(f"{text.replace('там', 'там там')=}")

print(f'{text=}')
text = text.replace('там', 'там там')
print(f'{text=}')

print(f"{text.replace('то', '', 1)=}") # Если выставить, 3-тий аргумент(количество замен) в replace, то замена происходит только определенное количество раз
````

`[str].strip()` - `[str]` очищает от пробельных символов в начале и в конце

````python
str_with_spaces = '                               строка с пробелами                       \n'
print(f'{str_with_spaces.strip()=}')
print(f'{str_with_spaces=}')

str_with_spaces = str_with_spaces.strip()
print(f'{str_with_spaces=}')
````

`[str].split(<разделитель> - по умолчанию пробел, <количество> -  по умолчанию до конца*)` - Разбивает строчку на отдельные части по `<разделителю>` определенное `<количество>` раз

````python
message = 'Пришло какое-то сообщение с каким-то определенным содержанием'
print(f'{ message= }')

print(f'{ message.split()= }')

print(f'{ message.split("-")= }')
print(f'{ message.split(sep="-")= }')

print(f'{ message.split(sep="-", maxsplit=1)= }')
print(f'{ message.split(maxsplit=3)= }')

words_in_message = message.split()
print(words_in_message, type(words_in_message))
````

`[str].count(<подстрока>, <начальный индекс, по умол. 0>, <конечный индекс, по умол. -1>)` - Подсчитывает количество `<подстрок>` в строке `[str]` на заданном интервале (`<начальный индекс>`, `<конечный индекс> - 1`)

````python
letter = 'Слово слово слово, за какие то слова'
print(f'{letter.count("С")=}')
print(f'{letter.count("с")=}')
print(f'{letter.count("ь")=}')

print(f'{letter.count("слово")=}')
print(f'{letter.count("слова")=}')
print(f'{letter.count("слов")=}')
# Поиск в конкретном промежутке строки 
print(f'{letter.count("слово", 0, 11)=} {letter[0: 11]=}')

start = 0
end = -1
find_str = 'лов'
print(f'{letter.count(find_str, start, end)=} {letter[start: end]=}')
````

`[str].find(<подстрока>, *<начальный индекс, по умол. 0>, *<конечный индекс, по умол. -1>)` - метод который ищет внутри строки указанную подстроку и возвращает индекс первого символа подстроки в строке. Если подстрока не найдена, метод возвращает -1

````python
letter1 = 'Слово слово слово, за какие то слова'
print(f'{letter1.find("С")=}')
print(f'{letter1.find("с")=}')
print(f'{letter1.find("ь")=}')
# Поиск в конкретном промежутке строки 
print(f'{letter.find("слово", 12, 20)=} {letter[0: 11]=}')
````

`[str].join(<список строк>)` - это метод который позволяет сделать слияние из списка строк в одну общую строку

````python
print(f'{words_in_message=}')
print(f'{"".join(words_in_message)=}')
print(f'{" ".join(words_in_message)=}')
print(f'{", ".join(words_in_message)=}')
````
