# Список - `[list]` - он же массив

Список -- изменяемый итерируемый(не один объект, а связка объектов, с которыми можно взаимодействовать) тип данных. 

## Cвойства

- Изменяемый
- Индексируемый
- Хранит любой тип данных

## Синтаксис: 

[ x0 ,x1, x2, x3, ..., xN] -- при этом x может быть любым объектом, любого типа данных.
[] - Пустой списко
## Индексация



```python
lst = ['e', 'l', 'e', 'c', 't', 'r', 'o', 'i', 'c', 'a']
print(lst[2:8])
print(lst[-9:-1])
```

```python
lst = []
lst = list()
lst2 = [2,3,4]
lst3 = [2, 'string', 56.65, True]
```

```python
print(lst2[1])
print(lst3[:])
print(lst3[:2])
print(lst3[2:])
print(lst3[2:100000000000000000000]) # Последнее значение индекса в срезе, может быть любым числов, но если вам нужно пройтись до конца не прописывайте никаких чисел
print(lst3[::2]) # - от начала и до конца мы будем идти с шагом 2
```

```python
print(f'{lst3=}')
lst3[2] = 50
print(f'{lst3=}')
```

# Методы списков

## Методы добавления

`[list].append(<то, что добавляем>)` - добавление в конец списка, к которому мы обращаемся, конкретно переданного элемента в скобки

```python
print(f'{lst=}')
print(f'{lst.append(1)=}')
lst.append(1)
lst.append(2)
lst.append(2)
lst.append(2)
print(f'{lst=}')
```

`[list].insert(<индекс куда всталяем значение>, <само значение\объект>)` - вставляет в спискок, к которому мы обращаеся, на определенную позицицию значение или объет

```python
print(f'{lst=}')
print(f'{lst.insert(2, 3)=} {lst=}') #[1, 1, 3, 2, 2, 2]
lst.insert(2, 3) #[1, 1, 3, 3, 2, 2, 2]
print(f'{lst=}')
```

`[list].extend(<[list2] второй список>)` - объединяет 2 списка в один общий, добавляя в конец обращаемого списка, добавляемый(в аргументе)

```python
print(f'{lst} | {lst2}')
print(f'{lst.extend(lst2)=}')
print(f"{lst=} | {lst2=}")
print(f'{lst2.extend(lst)=}')
print(f"{lst=} | {lst2=}")
```

## Методы удаления

`[list].remove(<значение\объект>)` - удаляем объект\значение из списка, но если их несколько, то только самое крайнее к левой границе. Если значений или значения уже в списке нет, то этот метод может вызвать ошибку

```python
print(f'{lst2=}')
print(f'{lst2.remove(2)=}')
print(f'{lst2=}')
```

`[list].pop(<Индекс элемента>)` -- "выдергивает" элемент(объект\значение) из списка, сохраняя его где-либо(по необходимости) 

```python
print(f'{lst2=}')
val = lst2.pop(1)
print(f'{val=} | {lst2=}')
print(f'{lst2.pop(1)=} | {lst2=}')
```

`[list].count(<объект\значение>)` -- ищет количество совпадающих с `<значение\объект>` элементов внутри обращаемого списка `[list]`

```python
print(f'{lst2=}')
print(f'{lst2.count(1)=}')
print(f'{lst2.count(2)=}')
print(f'{lst2.count(3)=}')
print(f'{lst2.count(4)=}')
```

`[list].reverse()` - Переворачивает элементы списка(зеркалит)

```python
print(f'{lst2=}')
lst2.reverse()
print(f'{lst2=}')
```

`[list].copy()` - создает клон списка `[list]`

```python
# неправильное копирование
# РАБОТАЕТ СО ВСЕМ ИЗМЕНЯЕМЫМИ ТИПАМИ ДАННЫХ
lst2_new = lst2 # это не копирование, это создание ссылки на объект / этот процесс называется перессылка
print(f"{lst2_new=}\n{lst2=}")
lst2.append('99999')
lst2_new.append('8888')
print(f"{lst2_new=}\n{lst2=}")
print(f"{id(lst2_new)=}\n{id(lst2)=}")
```

```python
lst2_new = lst2.copy() # создаем копию списка (отдельный объект)
# lst2_new = lst2
print(f"{lst2_new=} | {lst2=}")
print(f"{id(lst2_new)} | {id(lst2)}")
lst2_new.remove('99999')
print(f"{lst2_new=} | {lst2=}")
```

`[list].index(<объект\значение>)` - выводит индекс искомого `<объект\значение>` крайнего слева или ошибку `ValueError` если `<объект\значение>` в списке `[list]` нет

```python
print(lst, lst2)

print(lst.index(4)) # ищем индекс конкретного элемента - 4
print(lst3.index('string')) # ищем индекс конкретного элемента - string

finding_index = lst.index(2)
print(f'{finding_index=}')

finding_index = lst.index(1) # сохраняет индекс первого совпаденя слева
print(f'{finding_index=}')
# print(lst.index(5)) # попытка вывести индекс значения которого нет -- приведет к ошибке ValueError
```
